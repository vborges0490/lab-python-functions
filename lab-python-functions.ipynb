```python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        quantity = validate_inventory(f"Include the quantity for the product: {product} ")
        inventory[product] = quantity
    return inventory

def get_customer_orders():
    customer_orders = set()
    while True:
        product = validate_product("\nPlease inform the product that you want to add to the cart: ")
        customer_orders.add(product)
        more_product = input("Do you want to add more products? Write Yes or No: ").lower()
        if more_product == "no":
            break
    return customer_orders

def update_inventory(customer_orders, inventory):
    for product in customer_orders:
        if product in inventory:
            inventory[product] -= 1
    return inventory

def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders)
    percentage_of_unique_products_ordered = (total_products_ordered / len(products)) * 100
    return total_products_ordered, percentage_of_unique_products_ordered

def print_order_statistics(order_statistics):
    total_products_ordered, percentage_of_unique_products_ordered = order_statistics
    print("\n#### ORDER STATISTICS ####\n")
    print(f"Total Products Ordered: {total_products_ordered}")
    print(f"Percentage of Products Ordered: {percentage_of_unique_products_ordered}%\n")

def print_updated_inventory(inventory):
    print("#### NEW INVENTORY ####")
    for product, quantity in inventory.items():
        print(f"Product {product} has {quantity} products available")

def validate_inventory(value):
    while True:
        quantity = input(value)
        if quantity.isdigit():
            return int(quantity)
        print("This value is not valid.")

def validate_product(value):
    while True:
        product = input(value)
        if product in product_list:
            return product
        print("This product is not available! Try again!")

product_list = ["t-shirt", "mug", "hat", "book", "keychain"]

inventory = initialize_inventory(product_list)
customer_orders = get_customer_orders()
inventory = update_inventory(customer_orders, inventory)
order_statistics = calculate_order_statistics(customer_orders, product_list)
print_order_statistics(order_statistics)
print_updated_inventory(inventory)
```
